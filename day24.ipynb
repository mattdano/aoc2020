```python
store = []
for line in lines:
    lstore = []
    pos = 0
    while True:
        t = line[pos]
        if t=='e' or t=='w':
            lstore.append(t)
            pos+=1
        elif line[pos:pos+2] in ['se','sw','nw','ne']:
            lstore.append(line[pos:pos+2])
            pos+=2
        else:
            print(t,line[pos:pos+2])
        if pos>=len(line):
            break
    store.append(lstore)
        
```

```python
#lines[0],store[0]
```

```python
outcome = []
for moves in store:
    tilepos = complex(0)
    for move in moves:
        if move == 'e':
            tilepos+=1
        elif move == 'w':
            tilepos-=1
        elif move == 'nw':
            tilepos+= -1+1j
        elif move == 'ne':
            tilepos+= 0+1j
        elif move == 'sw':
            tilepos+= 0-1j
        elif move == 'se':
            tilepos+= +1-1j
    outcome.append(tilepos)
```

```python
directions = {"e":(1,0), "w":(-1,0), "ne":(0,1),  "nw":(-1,1), "se":(1,-1), "sw":(0,-1)}


hexgrid = dict()
for gridpos,flips in tally.items():
    # true is black
    x,y = (gridpos).real,(gridpos).imag
    hexgrid[(x,y)] = flips%2 == 1
    # add white neighbours
    for dirs in directions.values():
        dx,dy = dirs
        nexttile = (x+dx,y+dy)
        if not nexttile in hexgrid:
            hexgrid[nexttile] = False
    
print(len(hexgrid))


for i in range(100):
    flip = dict()
    for (x,y), colour in hexgrid.items():
        total = 0 
        for dirs in directions.values():
            dx,dy = dirs
            nexttile = (x+dx,y+dy)
            if nexttile in hexgrid and hexgrid[nexttile]:
                total +=1
        if colour and (total == 0 or total>2):
            flip[(x,y)] = False
        elif colour == False and total ==2:
            flip[(x,y)] = True
        
    for (x,y), colour in flip.items():
        hexgrid[(x,y)] = colour
        for dirs in directions.values():
            dx,dy = dirs
            nexttile = (x+dx,y+dy)
            if not nexttile in hexgrid:
                hexgrid[nexttile] = False

sum([h for h in hexgrid.values()])
```
