```python
f = open('./day3/input.txt')
lines = f.readlines()
```

```python
lines
```

```python
rights = [1,3,5,7]
tots = []

for r in rights:
    x = 0
    treecount=0
    for y,line in enumerate(lines):
        if y != 0:
            if line.replace('\n','')[x] == '#':
                treecount +=1
        x+=r
        x=x%len(line.replace('\n',''))
    tots.append(treecount)
    
x = 0
treecount=0
for y,line in enumerate(lines):
    if y != 0 and y%2==0:
        if line.replace('\n','')[x] == '#':
            treecount +=1
    if y%2 == 0:
        x+=1
        x=x%len(line.replace('\n',''))
tots.append(treecount)   
```
