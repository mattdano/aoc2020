```python
#How many bag colors can eventually contain at least one shiny gold bag? (The list of rules is quite long; make sure you get all of it.)

# idea
#make a directed graph. 
```

```python
import networkx as nx
```

```python
from collections import defaultdict
lookups = defaultdict(list)
all_cols = set()

def bag_parser(line):
    outer_bag_raw, inner_bags = line.split('contain')
    #print(outer_bag.replace('bags','').strip())
    outer_bag= outer_bag_raw.replace('bags','').strip()
    all_cols.add(outer_bag)

    for inner in inner_bags.split(','):
        #print(inner.strip())
        try:
            amt,colour=int(inner[0:2].strip()), inner[2:].replace('bags','').replace('bag','').replace('.','').strip()
        except ValueError:
            #print(line)
            if 'no other bags' in line:
                amt,colour=0, inner.replace('bags','').replace('bag','').replace('.','').strip()
                break
        #print(amt,',',colour)
        lookups[outer_bag].append([colour,amt])
        all_cols.add(colour)
        
for line in lines:
    bag_parser(line)
```

```python
G = nx.DiGraph()
G.add_nodes_from(all_cols)
```

```python
for k in lookups.keys():
    #print(lookups[k])
    for colour,amt in lookups[k]:
        G.add_edge(k,colour,count= amt)
    
```

```python
collect = set()
for col in all_cols:
    for path in nx.all_simple_paths(G, source=col, target='shiny gold'):
        print(col,path)
        for p in path:
            if p != 'shiny gold':
                collect.add(p)
```

```python
def get_bags(outbag):
    tally = 0
    if len(lookups[outbag]):
        for kv in lookups[outbag]:
            print(kv)
            tally += kv[1]
            tally += kv[1] * get_bags(kv[0])
        return tally
    else:
        return 0
get_bags('shiny gold')
```
